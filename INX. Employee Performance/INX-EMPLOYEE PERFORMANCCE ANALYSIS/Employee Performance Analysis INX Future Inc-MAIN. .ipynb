## 2. OUTLIER HANDLING

````python
# outlier checking for continuous variable
plt.figure(figsize=(15,20))                
plotnumber = 1
for i in num_data.columns :
    plt.subplot(5,6,plotnumber)
    sns.boxplot(data[i],color='r')
    plotnumber = plotnumber + 1
plt.tight_layout()
plt.show()
````

````python
### HANDLING OUTLIERS
def handle_outliers(df, column):
    # Calculate the 25th and 75th percentiles, and compute the IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace values outside the bounds with the respective bound
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# List of numerical columns to handle outliers for
columns_to_handle = [
    'TrainingTimesLastYear',
    'ExperienceYearsAtThisCompany',
    'ExperienceYearsInCurrentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager',
    'NumCompaniesWorked',
    'TotalWorkExperienceInYears'
]

# Applying outlier handling for each column and store the result back in 'data'
for column in columns_to_handle:
    data = handle_outliers(data, column)
````

````python
# outlier checking for contineous variable
plt.figure(figsize=(15,20))                
plotnumber = 1
for i in num_data.columns :
    plt.subplot(5,6,plotnumber)
    sns.boxplot(data[i],color='r')
    plotnumber = plotnumber + 1
plt.tight_layout()
plt.show()
````

### Hyperparameter Tuning for MLPClassifier
